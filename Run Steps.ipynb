## Step 4: Upload to s3

#### Outlets

```python
prefix='AT'
curr_date = '202102'
import pandas as pd
import ast
with open(f'{prefix}_{curr_date}/{prefix}_results.txt', 'r') as f:
    outlets = f.readlines()
parsed = []
for outlet in outlets:
    try:
        _outlet = ast.literal_eval(outlet)
        parsed.append(_outlet)
    except:
        print (outlet)
outlets = pd.DataFrame(parsed)
outlets['reviews_nr'] = outlets['reviews_nr'].astype(int)
print(len(outlets))
outlets.sample(5)
```

```python
outlets['website']
```

```python
prev_outlets = pd.read_parquet(f's3://dashmote-product/thuisbezorgd/202012/PT_outlet_information.parquet.gzip')
outlets = pd.concat([prev_outlets, outlets]).drop_duplicates('id_source', keep = 'first')
```

```python
outlets
```

```python
import numpy as np
def parse_rating(x): 
    try:
        return float(x)
    except:
        return np.nan
outlets['rating'] = outlets['rating'].apply(parse_rating)
```

```python
outlets.to_parquet(f's3://dashmote-product/thuisbezorgd/{curr_date}/{prefix}_outlet_information.parquet.gzip', engine = 'fastparquet', compression = 'gzip')
```
