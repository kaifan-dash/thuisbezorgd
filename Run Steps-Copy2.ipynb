## Step 4: Upload to s3

#### Outlets

```python
with open(f'{prefix}/{prefix}_results.txt', 'r') as f:
    outlets = f.readlines()
outlets = [ast.literal_eval(x) for x in outlets]
outlets = pd.DataFrame(outlets)
outlets['reviews_nr'] = outlets['reviews_nr'].astype(int)
print(len(outlets))
outlets.sample(5)
```

```python
outlets.to_parquet(f's3://dashmote-product/thuisbezorgd/{curr_date}/{prefix}_outlet_information.parquet.gzip', engine = 'fastparquet', compression = 'gzip')
```

#### Meals

```python
with open(f'{prefix}/{prefix}_meals.txt', 'r') as f:
    meals = f.readlines()
meals = [ast.literal_eval(x) for x in meals]
meals = pd.DataFrame(meals)
meals.sample(5)
```

```python
meals.to_parquet(f's3://dashmote-product/thuisbezorgd/{curr_date}/{prefix}_outlet_meal.parquet.gzip', engine = 'fastparquet', compression = 'gzip')
```
